```python
# Step 1: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Step 2: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Read the uploaded CSV file
import io
# Check if any file was uploaded
if uploaded:
    # Get the filename of the first uploaded file
    filename = next(iter(uploaded))
    df = pd.read_csv(io.BytesIO(uploaded[filename]))

    # Step 4: Show the DataFrame
    print("Original DataFrame:")
    print(df)

    # Task 1: Plot a line graph showing the number of news sold for each month
    plt.figure(figsize=(10, 5))
    plt.plot(df['month'], df['news_sold'], marker='o', linestyle='-', color='blue')
    plt.title('News Sold per Month')
    plt.xlabel('Month')
    plt.ylabel('News Sold')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Task 2: Add "sales_category" column based on condition
    df['sales_category'] = df['news_sold'].apply(lambda x: 'High' if x > 3000 else 'Low')
    print("\nDataFrame with sales_category column:")
    print(df)

    # Task 3: Sort DataFrame in descending order based on news_sold
    df_sorted = df.sort_values(by='news_sold', ascending=False)
    print("\nDataFrame sorted by news_sold (descending):")
    print(df_sorted)

    # Task 4: Calculate and print average number of news sold
    average_sales = df['news_sold'].mean()
    print("\nAverage number of news sold across all months:", average_sales)

    # Task 5: Find and display the month(s) with the maximum number of news sold
    max_sold = df['news_sold'].max()
    max_months = df[df['news_sold'] == max_sold]
    print("\nMonth(s) with maximum number of news sold:")
    print(max_months)
else:
    print("No file was uploaded. Please upload the CSV file.")
```
